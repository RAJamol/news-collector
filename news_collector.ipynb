```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.by import By
import time
```

```python
options = webdriver.ChromeOptions()
options.add_argument("--headless")
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()), options=options)
```

```python
KEYWORDS = [
    "lpg shortage",
    "gas shortage",
    "gas supply shortage",
    "cooking gas crisis",
    "hotel",
    "restaurant",
    "eatery",
    "hospitality"
]
```

```python
import requests
from bs4 import BeautifulSoup

def get_article_links():
    url = "https://news.google.com/rss/search?q=LPG+shortage+restaurant+hotel+gas+supply"
    
    response = requests.get(url)
    soup = BeautifulSoup(response.content, "xml")

    links = []
    for item in soup.find_all("item"):
        links.append(item.link.text)

    return links
```

```python
from newspaper import Article

def extract_article(url):
    try:
        article = Article(url)
        article.download()
        article.parse()
        return article.text
    except Exception as e:
        print(f"Failed: {url}")
        return None
```

```python
def is_relevant(text):
    text_lower = text.lower()
    return any(keyword in text_lower for keyword in KEYWORDS)
```

```python
def get_relevant_articles():
    links = get_article_links()
    results = []

    for url in links:
        driver.get(url)
        time.sleep(1)  # wait for redirects + page load

        # optional: get final resolved URL
        final_url = driver.current_url
        print(f"Processing: {final_url}")
        text = extract_article(final_url)

        if text and is_relevant(text):
            results.append({
                "url": final_url,
                "text": text  # preview
            })

    return results
```

```python
articles = get_relevant_articles()

```
